```python
import numpy as np
from sklearn.naive_bayes import GaussianNB
```

```python
# Features
X = np.array([
    [0, 0, 2, 0],
    [1, 1, 1, 1],
    [2, 0, 0, 0],
    [2, 2, 0, 1],
    [1, 2, 1, 0],
    [0, 1, 0, 0],
    [0, 2, 1, 0],
    [1, 0, 2, 1],
    [2, 2, 2, 1],
    [2, 0, 2, 1]
]) # Label
Y = np.array([1, 1, 0, 1, 0, 0, 0, 1, 0, 1]) # Label
```

```python
model = GaussianNB()
model.fit(X, Y)
```

Parantez içindeki sayıları istediğiniz gibi değiştirerek istediğiniz verilerin sonucuna bakabilirsiniz.

```python
print(model.predict([[2, 1, 2, 0]]))
```
